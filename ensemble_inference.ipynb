```python
from tf_utils.tf_data import _strict_data, overfitting_data
from class_inference import Single_Models, Models
from Inference_utils import split_into_single_list, model_batch_testing, vote_3_views, output_metrics_from_preds, vote_and_output, ensemble_votes
import tensorflow as tf
import numpy as np
from triple_data_utils import _show_array_of_images
import matplotlib.pyplot as plt
import pickle
```

## Get data

```python
val_x_triple, val_y_triple, test_x_triple, test_y_triple = _strict_data(1, 1, 1, 1, True)
print(val_x_triple.shape, val_y_triple.shape, test_x_triple.shape, test_y_triple.shape)
```

```python
new_val_x, new_val_y = split_into_single_list(val_x_triple, val_y_triple, 3)
new_test_x, new_test_y = split_into_single_list(test_x_triple, test_y_triple, 3)
```

```python
tot_cases = np.array((new_val_x + new_test_x), np.float32)
tot_labels = np.array((new_val_y + new_test_y), np.int64)
print(tot_cases.shape, tot_labels.shape)
```

## Load the best models pickle 

```python
with open('highest_trained_models.pkl', 'rb') as f:
    highest_trained_models = pickle.load(f)
```

```python
highest_trained_models.keys()
```

## Running the combined val/test set through each model type and parameter

```python
now_path = highest_trained_models['third_single_v3']
inception_models = model_batch_testing(now_path, Single_Models.Inception, tot_cases, tot_labels)
```

```python
now_path = highest_trained_models['third_resnet']
simple_resnet_models = model_batch_testing(now_path, Single_Models.simple_resnet, tot_cases, tot_labels)
```

```python
now_path = highest_trained_models['resnet_aux']
aux_resnet_models = model_batch_testing(now_path, Single_Models.resnet_aux, tot_cases, tot_labels)
```

```python
now_path = highest_trained_models['xception_1']
xception1_models = model_batch_testing(now_path, Single_Models.simple_Xception, tot_cases, tot_labels)
```

```python
now_path = highest_trained_models['xception_2']
xception2_models = model_batch_testing(now_path, Single_Models.simple_Xception, tot_cases, tot_labels)
```

```python
now_path = highest_trained_models['xception_aux']
xception_aux_models = model_batch_testing(now_path, Single_Models.Xception_aux, tot_cases, tot_labels)
```

## Ensembling

## 5 models

```python
for i, j in xception_aux_models.items():
    print('xception_aux_{}'.format(i))
    
    for k, l in aux_resnet_models.items():
        print('aux_resnet_{}'.format(k))
        
        for m, n in inception_models.items():
            print('inception_{}'.format(m))
            
            for o, p in simple_resnet_models.items():
                print('simple_resent_{}'.format(o))
                
                for q, r in xception1_models.items():
                    print('xception_{}'.format(q))
                    consolid_votes = ensemble_votes([j, l, n, p, r])
                    output_metrics_from_preds(consolid_votes, tot_labels)
                    vote_and_output(consolid_votes, tot_labels)
```
